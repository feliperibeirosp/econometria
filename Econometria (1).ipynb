IPCA <BR>
O Índice Nacional de Preços ao Consumidor Amplo (IPCA) é o principal indicador utilizado no Brasil para medir a variação dos preços de produtos e serviços consumidos pela população. Calculado pelo IBGE, o IPCA abrange despesas de alimentação, habitação, transportes, saúde, educação, entre outros setores, refletindo a inflação no país.

Atualizado em 09 Mai, 2025 15:18

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leitura do arquivo
data = pd.read_csv('indice_lista.csv', delimiter=';')

# Converter a coluna 'Mes-Ano' para o tipo datetime
data['Mes-Ano'] = pd.to_datetime(data['Mes-Ano'], format='%m-%Y')

# Criar o gráfico
plt.figure(figsize=(10,6))
plt.plot(data['Mes-Ano'], data['Indice'], marker='o', linestyle='-', color='b')

# Adicionar título e rótulos
plt.title('Índice IPCA')
plt.xlabel('Data (Mês-Ano)')
plt.ylabel('Índice')

# Exibir o gráfico
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import mplcursors as mpl

# Leitura do arquivo
data = pd.read_csv('pmc_novo.csv', delimiter=';')

# Converter a coluna 'Mes-Ano' para o tipo datetime
data['Mes-Ano'] = pd.to_datetime(data['Mes-Ano'], format='%m-%Y')

plt.figure(figsize=(10,6))
plt.plot(data['Mes-Ano'], data['Indice'], marker='o', linestyle='-', color='b')

# Adicionar título e rótulos
plt.title('Índice PMC')
plt.xlabel('Data (Mês-Ano)')
plt.ylabel('Índice')

# Adicionar interatividade com mplcursors
mpl.cursor(hover=True).connect(
    "add", lambda sel: sel.annotation.set_text(
        f'{data["Mes-Ano"].iloc[sel.index]}: {data["Indice"].iloc[sel.index]:.2f}'
    )
)

# Exibir o gráfico
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Para a série com tendência aplique o operador diferença e compare o resultado.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('pmc_novo.csv', delimiter=';')

# Converter a coluna 'Mes-Ano' para o tipo datetime
data['Mes-Ano'] = pd.to_datetime(data['Mes-Ano'], format='%m-%Y')

# Aplicando a diferenciação
data.set_index('Mes-Ano', inplace=True)

# Aplicando a diferenciação
data['Diferenca'] = data['Indice'].diff()

# Plotando os gráficos
plt.figure(figsize=(12, 6))

# Série com tendência
plt.subplot(1, 2, 1)
plt.plot(data.index, data['Indice'], label='Série com tendência')
plt.title('Índice do volume de vendas no comércio varejista (PMC)')
plt.xlabel('Data')
plt.ylabel('Índice')
plt.xticks(rotation=45)

# Série após a diferença
plt.subplot(1, 2, 2)
plt.plot(data.index[1:], data['Diferenca'].dropna(), label='Série após Diferença', color='orange')
plt.title('Série Após Diferença')
plt.xlabel('Data')
plt.ylabel('Diferenca do Índice')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
